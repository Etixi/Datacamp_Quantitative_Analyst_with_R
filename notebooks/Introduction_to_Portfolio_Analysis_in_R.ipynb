####**7. Rendement moyen arithmétique**
+ Les formules que je vais montrer supposent que nous avons un échantillon d’observations de rendement de portefeuille T: R1, R2, R3 jusqu’à RT. 

+ **Ensuite, le rendement moyen du portefeuille peut être estimé comme la valeur moyenne de ces observations de rendement T. C’est ce qu’on appelle aussi le rendement moyen arithmétique**. 

+ Il indique l’importance moyenne du rendement du portefeuille. 

+ La moyenne est importante : 
  + vous en gagnez, vous en perdez, mais en moyenne, le rendement du portefeuille doit être suffisamment élevé pour compenser le risque d’investissement.

####**8. Risque : volatilité du portefeuille**

+ Ce risque provient du fait que les rendements peuvent s’écarter du rendement moyen. 

+ J’utiliserai le nom « rendement déprécié » pour désigner la différence entre le rendement et sa valeur moyenne. 

+ Si le rendement rabaissé est positif, alors le rendement est supérieur à la moyenne. 

+ Si nous prenons la moyenne des rendements du portefeuille au carré, nous obtenons la variance du portefeuille. 

+ La variance a des valeurs importantes lorsque le rendement du portefeuille peut s’écarter beaucoup de sa valeur moyenne. 

+ Le plus souvent, nous n’utilisons pas la variance du portefeuille elle-même, mais sa racine carrée appelée écart-type du portefeuille, ou aussi, la volatilité du portefeuille. 

+ **Plus la volatilité est élevée, plus la probabilité d’un rendement positif ou négatif important de votre portefeuille est élevée.**

####**9. Pas de compensation linéaire en retour**
+ **La volatilité engendre donc des risques. Cela provoque également un décalage entre le rendement moyen et le rendement réel de l’investissement.** 

####**10. Pas de compensation linéaire en retour**
+ Supposons par exemple que l’investisseur réalise un gain de 50% et une perte de 50%. Alors la moyenne arithmétique de ces rendements est donc la moyenne de plus 50% et moins 50%, ce qui est nul. 

####**11. Pas de compensation linéaire en retour**

+ Le rendement moyen nul contraste avec le résultat réel pour l’investisseur, puisque, comme vous pouvez le voir sur la diapositive, la valeur finale de l’investissement n’est que de 75% de la valeur initiale. 

+ En raison de ce décalage entre le rendement moyen et le rendement réel des placements, les investisseurs préfèrent utiliser le rendement moyen géométrique.

####**12. Rendement moyen géométrique**
+ tenir compte du fait qu’il n’y a pas de compensation linéaire dans les rendements. 

+ Le rendement moyen géométrique est obtenu en multipliant d’abord les rendements totaux, puis en élevant ce nombre à la puissance de un divisé par le nombre d’observations. 

+ Si vous en soustrayez ensuite un de ce nombre, vous obtenez le rendement moyen géométrique. 

+ Dans notre exemple d’un rendement de plus 50 % et d’un rendement de moins 50 %, le rendement moyen géométrique est de moins 13,4 %, ce qui correspond, par définition, au rendement effectif moyen des placements.

###**Non-normalité de la distribution des retours**

####**1. Non-normalité de la distribution des retours**
+ **Jusqu’à présent, nous avons utilisé la volatilité du portefeuille comme mesure du risque.** 

####**2. La volatilité décrit le risque « normal »**

+ En gros, l’hypothèse sous-jacente est que le rendement du portefeuille a une distribution normale et donc que sa fonction de densité est en forme de cloche. 

+ Il est symétrique de sorte que les gains sont tout aussi probables que les pertes de même ampleur.

####**3. Non-normalité du retour**
+ **Pour de nombreux actifs financiers, cette hypothèse d’une distribution normale est fausse.** 

####**4. Non-normalité du retour**
+ Leur distribution a tendance à être biaisée vers la gauche avec des queues plus grosses que ceux d’une distribution normale.

####**5. Non-normalité du retour**
+ En conséquence, lorsque nous zoomons sur la queue gauche de l’histogramme des rendements financiers, nous constatons souvent qu’il y a plus de rendements négatifs extrêmes que ce qui est possible dans une distribution normale. 

+ Lorsque cela se produit, un investisseur ne se contente plus d’utiliser uniquement l’écart-type comme mesure du risque, mais devrait également utiliser une mesure du risque de baisse qui quantifie le risque de perdre de l’argent. 

+ Une telle mesure du risque de baisse se concentre sur le côté gauche de la distribution du rendement, au lieu de considérer la distribution complète.

####**6. Demi-écart de rendement du portefeuille**
+ **Un moyen simple de transformer l’écart-type en mesure du risque de baisse consiste à supprimer les rendements supérieurs à la moyenne.**

####**Pilote 3 : La corrélation entre les rendements des actifs**

+ ***Le troisième facteur de performance du portefeuille est la corrélation entre les rendements des actifs. De manière générale, la corrélation vous indique comment deux rendements d'actifs ont tendance à évoluer ensemble.***

+ **La corrélation des actifs a des conséquences importantes sur la performance globale du portefeuille. Cette corrélation est importante car elle peut réduire la volatilité grâce à la diversification ou à la réduction de la corrélation globale. En fait, plus la corrélation est faible, plus le portefeuille a tendance à réussir à compenser partiellement les pertes importantes d'un actif avec seulement une perte mineure, voire un gain, dans un autre actif.**

+ **Dans le cas extrême de deux rendements d'actifs identiques, la corrélation sera de 1 et il n'y a pas de potentiel de diversification. Dans l'autre cas extrême où, si un rendement d'actif est supérieur à la moyenne et l'autre presque toujours inférieur à la moyenne, la corrélation est négative. La corrélation est de 0 lorsque les rendements des actifs sont linéairement indépendants les uns des autres. Notez que l'interdépendance peut toujours exister à un niveau non linéaire même lorsque la corrélation est de 0.**

###**Utilisation de la notation matricielle**

####**1. Utilisation de la notation matricielle**
+ ***Nous venons de voir les facteurs de moyenne et de variance du portefeuille dans le cas d’un portefeuille avec deux actifs. Dans cette vidéo, je vous montre comment ces résultats peuvent être étendus au cas général de $N$ actifs, où $N$ peut être n’importe quel nombre d’actifs.***

####**2. Variables en jeu pour n actifs**

+ Lorsque nous travaillons avec de nombreux actifs, nous devons simplement faire attention aux nombreuses variables qui entrent en jeu. 

+ Je vais vous montrer comment l’utilisation de matrices dans R vous aidera à garder l’analyse des données bien organisée. 

+ La première matrice que nous utilisons est la matrice des pondérations de portefeuille. 

+ Nous avons $N$ actifs et donc $N$ pondérations de portefeuille: $w_1$, $w_2$, jusqu’à $w_N$. 

+ Nous allons les empiler dans une matrice colonne de dimension $N$ fois 1 et l’appeler la matrice $w$. La matrice suivante que nous utilisons est la matrice des rendements des actifs. 

+ Pour chacun des actifs, nous avons un rendement, que nous notons $R_1$, $R_2$ jusqu’à $R_N$. 

+ Encore une fois, ces retours peuvent être empilés dans une matrice colonne de dimension $N$ fois 1 que j’appelle $R$. Pour chaque actif, nous avons également un rendement attendu. Je les désignerai comme $\mu_i$. 

+ Donc, $\mu_1$ est le rendement attendu de l’actif 1, $\mu_2$ est le rendement attendu de l’actif 2, et ainsi de suite. 

+ L’empilement des $N$ rendements attendus dans la matrice colonne $\mu$ nous donne la matrice des rendements attendus $\mu$. 

+ Enfin et surtout, il y a la matrice de covariance. Pour chacun des actifs, nous avons la variance de leur rendement, que je désigne par sigma carré $i$. 

+ Pour chaque paire de rendements d’actifs $i$ et $j$, nous avons une covariance, que je désigne par $\sigma_{ij}$. 

+ Nous avons donc $N$ variances, et pour chaque paire d’actifs, nous avons une covariance. 

+ Afin de gérer toutes ces variances et covariances, il est utile de les rassembler dans la matrice dite de covariance. 

+ Il s’agit d’une matrice carrée avec $N$ lignes et $N$ colonnes. Les éléments de la diagonale sont les variances de chacun des retours. 

+ En dehors de la diagonale se trouvent les covariances. Par exemple, à la ligne 2, colonne 1, nous avons la covariance entre le rendement de l’actif 2 et le rendement de l’actif 1.

+ Il s’ensuit que le rendement attendu du portefeuille est la moyenne pondérée des rendements attendus de tous les actifs $N$.

####**Calcul de la frontière efficiente à l'aide d'une grille de rendements cibles**

+ ***Comme vous l'avez vu, une approche pour calculer la frontière efficiente consiste à définir d'abord la grille des rendements cibles, puis, pour chaque rendement cible, à trouver le portefeuille dont le rendement attendu est égal au rendement cible à la variance la plus faible possible.***

+ **Mais qu'est-ce qu'une grille raisonnable de rendements cibles ? Vous définirez le rendement cible maximum sur le rendement moyen maximum des actions. Idéalement, vous fixeriez le rendement cible minimum au rendement du portefeuille à variance minimale. Comme vous ne connaissez pas encore ce rendement de portefeuille à variance minimale, vous créez une grille en utilisant le minimum des rendements moyens de toutes les actions.**

+ *Dans cet exercice, vous utiliserez une boucle for pour calculer votre grille de moyennes, d'écarts et de pondérations potentiels de portefeuille.*

####**Instructions**

+ Calculez les moyennes de colonne des rendements (en utilisant colMeans()) et appelez ce stockmu.
+ Créez une séquence (seq()) de longueur 50, qui commence à un taux sans risque de 1 % et se termine à la valeur maximale de stockmu appelée grille.
+ Initialisez deux vecteurs vides avec la même longueur que la grille à l'aide de rep(), où vous stockerez les moyennes et les écarts-types du portefeuille. Appelez-les vpm et vpsd.
+ Initialiser une matrice vide de 50 lignes et 30 colonnes. 
+ Appelez cela mweights. Vous pouvez utiliser la fonction matrix() pour le faire.
+ Créez une boucle for qui commence à la première valeur de grid et se termine à la dernière. 
+ La boucle for devrait créer un portefeuille appelé opt using return et avec un rendement cible de grille.
+ A chaque itération, la boucle for doit remplir 
  + les vecteurs vpm ($pm), 
  
  + vpsd ($ps) avec leurs valeurs respectives de opt.

+ Stockez les pondérations du portefeuille ligne par ligne en mweights ($pw).

```r
# Calculate each stocks mean returns
stockmu <- colMeans(returns)

# Create a grid of target values
grid <- seq(from = 0.01, to = max(stockmu), length.out = 50)

# Create empty vectors to store means and deviations
vpm <- vpsd <- rep(NA, length(grid))

# Create an empty matrix to store weights
mweights <- matrix(NA, 50, 30)

# Create your for loop
for(i in 1:length(grid)) {
  opt <- portfolio.optim(x = returns, pm = grid[i])
  vpm[i] <- opt$pm
  vpsd[i] <- opt$ps
  mweights[i, ] <- opt$pw
}
```

Dans cet exercice, vous utiliserez window() pour créer deux échantillons : un échantillon d'estimation et un échantillon d'évaluation. Cet exercice illustrera comment les pondérations de portefeuille peuvent différer lors du changement de la fenêtre d'estimation.***

+ **Pour vous rappeler, la fonction window() a pour argument x, start et end. Où début et fin sont au format "AAAA-MM-JJ".**

+ *L'objet renvoyé est chargé dans votre espace de travail.*

####**Instructions**

+ Créez l'exemple return_estim en sous-ensemble des retours, où l'échantillon commence le 1er janvier 1991 et se termine le 31 décembre 2003.
+ Créez l'exemple de retours_eval en créant des sous-ensembles de retours, où l'échantillon commence le premier jour de 2004 et se termine le dernier jour de 2015.
+ Créez un vecteur de poids maximum égal à 10 %, avec une longueur égale au nombre de colonnes qu'il y a dans les retours appelés max_weights.
+ Créez un portefeuille avec l'échantillon d'estimation appelé pf_estim, où le poids maximum (reshigh) est défini sur max_weights.
+ Créez un portefeuille avec l'échantillon d'évaluation appelé pf_eval, où le poids maximum (reshigh) est défini sur max_weights.
+ Créez un nuage de points des pondérations du portefeuille d'évaluation par rapport aux pondérations du portefeuille d'estimation (notez que vous pouvez utiliser $pw). 
+ Si les pondérations du portefeuille sont identiques, elles doivent se situer sur la ligne à 45 degrés.

```r
# Create returns_estim 
returns_estim <- window(returns, start = "1991-01-01", end = "2003-12-31")

# Create returns_eval
returns_eval <- window(returns, start = "2004-01-01", end = "2015-12-31")

# Create vector of max weights
max_weights <- rep(0.1, ncol(returns))

# Create portfolio with estimation sample 
pf_estim <- portfolio.optim(returns_estim, reshigh = max_weights)

# Create portfolio with evaluation sample
pf_eval <- portfolio.optim(returns_eval, reshigh = max_weights)

# Create a scatter plot with evaluation portfolio weights on the vertical axis
plot(pf_estim$pw, pf_eval$pw)
abline(a = 0, b = 1, lty = 3)
```

***Vous avez presque fini! En créant deux échantillons à l'aide de la fonction window() vous pouvez backtester, ou évaluer la performance d'un portefeuille sur des données historiques !***
