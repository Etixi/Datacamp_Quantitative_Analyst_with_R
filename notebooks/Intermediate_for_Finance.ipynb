####**1. Opérateurs logiques**

####**2. Opérateurs logiques**
+ Les opérateurs logiques sont là pour vous aider. Il existe trois opérateurs logiques.

####**3. Opérateurs logiques**
+ Et, représenté par une esperluette, est une intersection, ce qui signifie qu’il ne retournera TRUE que si les conditions de chaque côté de l’esperluette sont toutes deux vraies.

####**4. Opérateurs logiques**
+ Ou, représenté par une ligne verticale, ou tuyau, est une union. 
+ Cela signifie qu’il retournera TRUE si une seule des conditions est vraie, ou si les deux sont vraies.

####**5. Opérateurs logiques**
+ Non, comme vous l’avez vu, est représenté par un point d’exclamation, et cela annule, ou inverse, le signe logique d’une condition.

####**6. Opérateurs logiques**

+ Si vous avez une condition qui retourne TRUE et que vous mettez un not devant, elle sera inversée en FALSE.

####**7. Exemples d’opérateurs logiques**
+ Les opérateurs logiques ont beaucoup plus de sens après quelques exemples. 

+ Regardons-en quelques-uns. 
+ Le stock fictif de datacamp est actuellement au prix de 64 points-69. 
+ Supposons que vous vouliez savoir si cela se situe entre 64 et 65. 
+ De toute évidence, c’est le cas, mais comment pouvez-vous dire à R de vérifier cela? 

+ En d’autres termes, vous voulez vérifier si datacamp est supérieur à 64, ET datacamp est inférieur à 65. 

+ En traduisant cela directement en code, vous pouvez écrire ce qui suit. Sur le côté gauche de l’esperluette, la condition sera évaluée à TRUE car le stock de datacamp est supérieur à 64. 

+ Le côté droit évaluera également à TRUE car datacamp est inférieur à 65. 

+ Étant donné que les deux côtés de l’opérateur AND sont vrais, l’instruction entière sera évaluée à true. 

+ Les deux conditions n’ont pas besoin d’être sur le même objet. 

+ Si vous voulez savoir si datacamp est supérieur à 63 OU Apple est supérieur à 126, vous pouvez écrire cela comme ceci. 

+ Datacamp est supérieur à 63, donc le côté gauche est vrai, mais l’action Apple n’est pas supérieure à 126, donc le côté droit est faux. 

+ Cependant, puisque vous avez utilisé OR, seul 1 côté doit évaluer à true, donc cette expression est vraie.

 + Une instruction if est votre façon de dire à R, si cette condition est vraie, exécutez ce code, sinon la condition est fausse, alors ignorez-la et passez à autre chose.

####**3. Par défaut si...**
+ Voici un exemple. 

+ Soit les nombres compris entre 0 et 1 la probabilité qu’une entreprise fasse défaut. 

+ Pour générer ce nombre, nous utiliserons runif pour obtenir un nombre aléatoire compris entre 0 et 1. 

+ Vous ne savez pas quel est ce nombre à l’avance, mais vous voulez planifier en conséquence! 

+ Plus précisément, si la chance est au-dessus du point 5, vous souhaitez que votre programme vous avertisse en imprimant « Susceptible de faire défaut! ». 

+ Une instruction if est parfaite pour ce genre de logique. 

+ À l’intérieur de la parenthèse, vous utilisez l’un des opérateurs relationnels que vous avez appris précédemment pour créer la condition, et à l’intérieur des accolades, vous ajoutez une instruction print. 

+ Lors de cette exécution, l’instruction print a été exécutée. 

+ Cela doit signifier que default_chance était au-dessus du point 5, et bien sûr, c’était le point-948. 

+ Alors, que s’est-il passé ici? R a d’abord vérifié si le default_chance était au-dessus du point 5, et c’était le cas, alors il a exécuté le code à l’intérieur des accolades, imprimant l’instruction. 

+ Si nous exécutons à nouveau ce code avec une nouvelle chance par défaut, l’instruction print ne s’exécute pas. 

+ Cela doit signifier que le nouveau nombre aléatoire généré était inférieur au point 5 et, bien sûr, à son point 395.

####**4. Si-sinon**
+ Au lieu de ne rien faire si le default_chance est inférieur au point 5, vous pouvez ajouter plus de logique à l’aide d’une instruction if else afin que votre programme exécute un morceau de code différent si la condition est false. Cela ressemble à ceci. 

+ Vous pouvez lire ceci comme, si cette condition est vraie, exécutez le premier morceau de code, sinon, exécutez l’autre bloc. 

+ L’organigramme montre la logique. Tout d’abord, une condition de test est vérifiée et, en fonction du résultat, le corps de l’instruction if ou le corps de l’instruction else est exécuté.

####**5. Par défaut si... autre...**

+ Pour prolonger notre exemple, réglons simplement le default_chance sur le point 4. 

+ Une instruction else est ajoutée avec sa propre instruction print, et lorsque le programme est exécuté à nouveau, il imprime « Pas de problèmes ici! », ce qui est la réponse appropriée puisque default_chance est en dessous du point-5.

####**6. Si-sinon si-sinon**
+ Enfin, vous pouvez ajouter une deuxième condition avec la logique de si, sinon si, si.

+ Si la première condition est vraie, le premier segment de code est exécuté, si la première condition est fausse, mais la deuxième condition est vraie, le deuxième segment est exécuté. 

+ Et si les deux conditions sont fausses, le code de l’instruction else est exécuté. Gardez à l’esprit que même si nous n’avons utilisé que 2 déclarations if ici,

####**7. Par défaut si... sinon... autre**

###**vapply() - spécifiez votre sortie!**

####**1. vapply() - spécifiez votre sortie!**

+ Comme vous venez de le voir, lorsque sapply ne parvient pas à simplifier, il renvoie une liste tout comme lapply, sans aucun avertissement. 

+ Si vous l’utilisez à l’intérieur d’une fonction qui s’attend à ce que sapply renvoie un vecteur, cela pourrait faire des ravages et être pénible à déboguer lorsque cela ne fonctionne pas. 

+ Au lieu de cela, il existe une version plus robuste de sapply appelée vapply.

####**2. vapply()**


+ La robustesse vient d’un argument supplémentaire à appliquer appelé FUN-dot-VALUE. 

+ Vous spécifiez cette méthode à l’aide d’une combinaison du type d’objet qui doit être renvoyé à partir de chaque appel vapply, ainsi que de la longueur de l’objet. 

+ De manière compacte, cela peut être écrit comme, par exemple, character(1), en disant que vous vous attendez à ce que chaque appel vapply renvoie un vecteur de caractère de longueur 1, ou die en essayant. 

+ C’est ce que nous ferions ici avec notre exemple de liste de stock. 

+ Nous nous attendons à ce que chaque appel vapply renvoie un vecteur de caractères de longueur 1. 

+ Si ce n’est pas ce que get est retourné, vapply échoue et génère une erreur, ce qui est exactement ce que vous voulez. 

+ S’il réussit, il simplifie la sortie comme sapply. Ici, vous pouvez voir que cela a fonctionné comme prévu. 

+ Dans les exercices, vous verrez ce qui se passe lorsque vapply échoue.
