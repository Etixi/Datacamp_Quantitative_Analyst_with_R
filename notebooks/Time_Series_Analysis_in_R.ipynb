####**1. Fréquence d’échantillonnage**

####**2. Fréquence d’échantillonnage: exacte**

+ **Certaines données de séries chronologiques sont exactement espacées uniformément. Par exemple, des mesures horaires de température pour chaque heure d’une journée.**

####**3. Fréquence d’échantillonnage : approximative**

+ **Certaines données de séries chronologiques ne sont que peu régulièrement espacées. Par exemple, les mesures de température enregistrées chaque fois que vous consultez votre courrier électronique.**

####**4. Fréquence d’échantillonnage : valeurs manquantes**

+ **Certaines données de séries chronologiques sont régulièrement espacées, mais avec des valeurs manquantes. Par exemple, des mesures de température toutes les heures pendant que vous êtes éveillé.**

####**5. Hypothèses de base**

+ L’analyse des données de séries chronologiques se poursuit avec quelques hypothèses simplificatrices :

  + **La première hypothèse est que les observations consécutives sont également espacées.** 
  
  + **Deuxièmement, un indice d’observation en temps discret est appliqué**. 
  
+ **Dans la pratique, cela peut ne s’appliquer qu’approximativement, et parfois des données peuvent être manquantes.** 

+ **Par exemple, les retours journaliers sur un stock peuvent n’être disponibles que pendant les jours de semaine, et les données peuvent ne pas être disponibles pour certains jours fériés.** 

+ **Les valeurs mensuelles de l’IPC sont également espacées par mois, mais pas par jours.**

####**6. Fréquence d’échantillonnage: fonctions R**

+ **Vous pouvez appliquer la fonction start() à la série de mesures de température horaires pour confirmer qu’elle commence le premier jour à la première heure.** 

+ **De même, l’application de la fonction end() confirme que la dernière observation de la série a lieu le premier jour à l’heure 24.** 

+ **La fonction frequency() indique que 24 observations sont faites chaque jour, et la fonction deltat() note que les observations sont faites tous les 0-point-0417 jours, c’est-à-dire que l’incrément de temps entre les observations est de 1 sur 24.**

####**Quand la fréquence d’échantillonnage est-elle exacte?**

+ ***La fréquence d’échantillonnage n’est souvent qu’approximative et l’intervalle entre les observations n’est pas tout à fait fixe***. 

+ **Par exemple, il y a généralement 365 jours dans une année basée sur le calendrier grégorien. Cependant, (presque) tous les quatre ans, il y a 366 jours (années bissextiles). Cela compense le fait que la Terre effectue une rotation autour de Sol, le soleil, en environ 365,2422 jours, en moyenne.**

+ **En guise d’hypothèse simplificatrice, nous ignorons souvent ces petites divergences et procédons comme si la fréquence d’échantillonnage et les intervalles d’observation étaient des constantes fixes. En règle générale, nos résultats ne seront pas sensibles à l’approximation lorsque le processus sous-jacent ne change pas trop rapidement.**

+ *Lequel des éléments suivants a une fréquence d’échantillonnage exacte?*

####**Réponds à la question**


+ Observations mensuelles enregistrées le dernier jour du mois depuis plusieurs années


+ Observations hebdomadaires enregistrées tous les vendredis depuis plusieurs années


+ Observations quotidiennes enregistrées tous les « jours ouvrables » pendant plusieurs semaines

+ ***Observations horaires enregistrées chaque heure pendant plusieurs jours***

***C'est exact! Il y a toujours exactement le même nombre d'heures dans chaque jour sans exception.***

+ Les tendances à la hausse peuvent augmenter plus rapidement que linéairement.

+ La décroissance rapide est également une possibilité, mais elle n’est pas aussi courante dans la plupart des applications.

+ **Certaines séries peuvent présenter des tendances périodiques ou sinusoïdales au fil du temps**.

+  Dans la figure A, vous voyez une série périodique avec une durée de cycle d’environ 75 observations. 

+ Dans la figure B, la série oscille plus rapidement et la durée du cycle est beaucoup plus petite.

+ Les séries chronologiques peuvent également présenter des tendances en matière de variabilité.

+  Les figures A et B montrent toutes deux des exemples de séries présentant des tendances à la hausse de la variance au fil du temps.

####**7. Exemples de transformations : log()**

+ **Discutons maintenant de quelques transformations de données simples qui peuvent être utilisées pour supprimer ou filtrer certaines des tendances courantes dont nous venons de parler**. 

+ **La fonction log() dans R, qui est la transformation logarithme naturelle, peut linéariser une tendance de croissance rapide.** 

+ **Vous pouvez voir que la croissance rapide à gauche est maintenant une tendance temporelle linéaire à droite.** 

+ **La transformation logarithmique peut également stabiliser une série qui présente une variance croissante.** 

+ **La principale restriction est que la transformation logarithmique n’est définie que pour les séries chronologiques à valeur positive.**

####**8. Exemples de transformations : diff()**

+ **La fonction diff() dans R, qui est la première transformation de différence, peut supprimer les tendances linéaires.** 

+ Vous pouvez voir que la tendance linéaire qui apparaît dans la figure A est éliminée dans la figure B. 

+ La série de la figure B est appelée série de différences ou série de changements. 

+ **Ses valeurs représentent les incréments ou les changements entre les observations consécutives de la série originale**. 

+ **Bien sûr, une première série de différences aura toujours une observation de moins que la série originale**.

####**9. Exemples de transformations : diff(..., s)**

+ **Enfin, pour les séries périodiques, la fonction diff() dans R avec l’argument saisonnier supplémentaire S, peut supprimer les tendances périodiques**. 

+ **C’est ce qu’on appelle la transformation des différences saisonnières**. 

+ **La série de la figure A présente un schéma périodique ou saisonnier avec une durée de cycle 4. Après avoir appliqué la différenciation saisonnière avec S égal à 4, nous voyons que la série de la figure B ne présente aucune tendance périodique.**

###**Le modèle de marche aléatoire (RW)**

####**2. Marche aléatoire**

+ ***Une marche aléatoire ou R-W est un exemple simple de processus instable ou non stationnaire. Il n’a pas de moyenne ou de variance spécifiée. ####**3. Marche aléatoire**

####**4. Marche aléatoire**

+ Regardons maintenant le modèle de marche aléatoire. 

+ Il est défini récursivement. L’observation d’aujourd’hui ou actuelle est égale à l’observation d’hier ou de précédente, plus bruit ou erreur.

+ Souvent, la valeur zéro est utilisée comme point initial $Y_0$, pour plus de simplicité.

+ **Ici, vous voyez que la différence entre les observations actuelles $Y_t$ et l’observation précédente $Y_{t-1}$ est égale au terme bruit. C’est-à-dire que la différence ou la série de changement est le bruit blanc.**

####**1. Processus stationnaires**

####**2. Stationnarité**

+ ***Pour obtenir la parcimonie dans un modèle de séries chronologiques, vous supposez souvent une certaine forme d’invariance distributive au fil du temps, ou stationnarité.*** 

+ ***Pour une série chronologique observée : Les fluctuations semblent aléatoires***. 

+ Cependant, le même type de comportement aléatoire est souvent valable d’une période à l’autre. 

+ Par exemple, les rendements des actions ou les variations des taux d’intérêt ont chacun un comportement très différent de celui de l’année précédente. 

+ **Mais leur moyenne, leur écart-type et d’autres propriétés statistiques sont souvent similaires d’une année à l’autre.**

####**3. Stationnarité faible - I**

+ ***Un processus est faiblement stationnaire si sa moyenne, sa variance et sa covariance sont inchangées par les décalages temporels. Autrement dit, il existe une constante commune mu et une variance sigma-carré pour tout temps $t$.***

####**4. Stationnarité faible - II**

+ Et, la covariance entre $Y$ à l’instant $T$ et à l’instant $S$ ne dépend que de la proximité de $T$ et $S$, et non des indices de temps $T$ et $S$ eux-mêmes. Par exemple, si $Y$ est un processus stationnaire, alors la covariance entre $Y$ aux temps 2 et 5 est la même que celle entre les temps 7 et 10, puisqu’ils sont tous deux séparés de trois unités de temps.

####**5. Stationnarité : pourquoi ?**

+ Pourquoi se concentrer sur les modèles stationnaires ? Un processus stationnaire peut être modélisé avec beaucoup moins de paramètres. Vous n’avez pas besoin d’une moyenne différente pour chaque $Y_t$; Au contraire, ils ont tous un mu moyen commun. Vous pouvez estimer mu avec précision par y-bar, la moyenne de l’échantillon.

####**6. Stationnarité : quand ?**

+ **Lorsqu’une série chronologique est observée, une question naturelle est : est-elle stationnaire ?**

+ **Plus précisément, un modèle de processus stationnaire est-il approprié pour ces données de séries chronologiques?** 

+ **De nombreuses séries chronologiques financières ne présentent pas de stationnarité. Toutefois: Les changements dans la série sont souvent approximativement stationnaires.** 

+ **Peut-être après avoir appliqué une transformation log(). Une série stationnaire devrait montrer une oscillation aléatoire autour d’un niveau fixe, un phénomène appelé réversion moyenne.**

###**Le modèle autorégressif**


#####**1. Le modèle autorégressif**

+ ***Il existe de nombreux processus autorégressifs ou AR.***

####**2. Le modèle autorégressif - I**

+ **Vous vous concentrerez sur le cas le plus simple, de premier ordre, dans lequel l’observation d’aujourd’hui est régressée sur l’observation d’hier, à tout moment $t$. **

+ **Dans R, vous travaillerez avec la version centrée moyenne du modèle. Aujourd’hui moins la moyenne est égal à un paramètre de pente temps hier moins la moyenne, plus le bruit.**

####**3. Le modèle autorégressif - II**

+ ***Plus formellement, si epsilon est un bruit blanc nul, alors $Y$ est un processus AR s’il suit l’équation affichée, pour certains paramètres constants mu et phi. Mu est la moyenne, Phi est le coefficient de pente, et il y a un troisième paramètre, sigma-carré, la variance du bruit blanc.**

####**4. Processus de AR - I**

+ Si la pente phi est nulle, Y est simplement un processus de bruit blanc. 

+ Si la pente phi n’est PAS nulle, alors $Y$ dépend à la fois du bruit actuel et de l’observation précédente, et le processus $Y$ est autocorrélé. 

+ De grandes valeurs de la pente phi conduisent à une plus grande autocorrélation. 

+ Et comme vous le verrez sur la diapositive suivante, les valeurs négatives de la pente phi donnent lieu à des séries temporelles oscillatoires.

####**5. Exemples de AR**

+ Maintenant, regardons quatre exemples de séries chronologiques AR. 

+ En haut à gauche, vous voyez une série AR avec phi égal à 0-point-98, et en dessous, un avec phi égal à 1. 

+ Dans les deux cas, la série montre une forte persistance dans le niveau, ce qui signifie que chaque observation est proche de ses voisines. 

+ En haut à droite, vous voyez une série AR avec phi égal à -0-point-6. 

+ Les séries AR avec des coefficients de pente négatifs présenteront des modèles oscillatoires, comme vous le voyez ici. 

+ Enfin, en bas à droite, vous voyez une série AR avec phi égal à 1-point-01. Ici, il y a tellement de dépendance que la série diverge rapidement, vers le bas dans ce cas.

####**6. Autocorrélations**

+ ***Le processus AR a une autocorrélation. Ceci est déterminé par le paramètre de pente phi. Dans les trois premiers panneaux, vous pouvez comparer la forme de la fonction d’autocorrélation pour différentes valeurs positives de phi, 0 virgule 95, 0 point 75 et 0 point 2, respectivement.***

+ **Notez une décomposition plus lente lorsque phi est plus gros**.

####**7. Marche aléatoire**

###**Le modèle de la moyenne mobile simple**

####**1. Le modèle de la moyenne mobile simple**

+ ***Les modèles avec autocorrélation peuvent être construits à partir du bruit blanc.***

####**2. Le modèle de la moyenne mobile simple**

+ Une somme pondérée du bruit actuel et du bruit précédent est appelée processus de moyenne mobile simple. 

+ Il existe de nombreux processus de moyenne mobile ou MA. 

+ Vous vous concentrerez sur le cas le plus simple, de premier ordre, dans lequel l’observation d’aujourd’hui est régressée sur le bruit d’hier. 

+ Si epsilon est un bruit blanc moyen nul, alors $Y$ est un processus MA s’il suit l’équation affichée. 

+ Il y a trois paramètres: **Mu est la moyenne, Theta est le paramètre de pente et sigma-carré est la variance du bruit blanc.**

####**3. Processus d’AM - I**
+ **Si le paramètre de pente thêta est nul, $Y$ est simplement un processus de bruit blanc.**

####**4. Processus d’AM - II**
+ Si le paramètre de pente thêta n’est PAS nul, $Y$ dépend à la fois du bruit actuel et du bruit précédent, et le processus $Y$ est autocorrélé. 

+ De grandes valeurs de la pente thêta conduisent à une plus grande autocorrélation. Et les valeurs négatives de la pente thêta entraînent des séries temporelles oscillatoires.

####**5. Exemples d’AM**
+ Examinons maintenant quatre exemples de séries chronologiques MA. 

+ En haut à gauche, vous voyez une série MA avec thêta égal à 0-point-5, et à sa droite, une avec thêta égal à 0-point-9. 

+ Dans les deux cas, la série montre une persistance dans le niveau, ce qui signifie que chaque observation est proche de ses voisines. 

+ Cette persistance est plus prononcée à droite pour la plus grande valeur de thêta. 

+ En bas à gauche, vous voyez une série MA avec thêta égal à -0-5-point-0. 

+ Les séries MA avec des coefficients MA négatifs ont tendance à présenter des modèles alternatifs ou oscillatoires, comme vous le voyez ici. 

+ Enfin, en bas à droite, vous voyez une série MA avec thêta égal à <>. Ici, il n’y a pas d’autocorrélation. Rappelez-vous, dans ce cas, le processus est un bruit blanc.

####**6. Autocorrélations**
+ Le processus MA a une autocorrélation déterminée par le paramètre de pente thêta, cependant, il n’a de dépendance que pendant une période. 

+ L’autocorrélation est nulle à des décalages de 2 et plus. Si thêta est positif, l’autocorrélation décalage 1 sera positive, si thêta est négatif, l’autocorrélation décalage 1 sera négative.

###**Comparez les modèles AR et MA**
####**1. Comparez les modèles AR et MA**

####**2. Processus d’MA et de AR**

+ Les modèles MA et AR sont similaires à bien des égards. 

+ Les deux ont des termes de bruit blanc moyens zéro epsilon avec un paramètre de variance sigma-carré. 

+ Les deux incluent un paramètre moyen mu. 

+ Dans le modèle MA, le $Y_t$ d’observation d’aujourd’hui est régressé sur le bruit d’hier $\epsilon_t$ moins 1. 

+ Dans le modèle AR, le $Y_t$ d’observation d’aujourd’hui est régressé sur l’observation d’hier $Y_t$ moins 1. Cependant, le modèle MA n’a d’autocorrélation qu’à un seul décalage, mais le modèle AR a une autocorrélation à de nombreux décalages.

####**3. Processus MA et AR : autocorrélations**

+ Considérons un modèle MA avec thêta égal à 0 point 75. 

+ Son autocorrélation de décalage 1 est 0-point-48, avec une autocorrélation nulle à d’autres décalages, comme indiqué en haut à gauche. 

+ Un modèle AR avec phi égal à 0-point-48 a également un décalage 1 autocorrélation 0-point-48. 

+ Cependant, son autocorrélation diminue à zéro à des décalages plus élevés, comme indiqué en haut à droite. 

+ Un modèle MA avec thêta égal à -0-point-75 a un décalage de 1 autocorrélation -0-point-48, et une autocorrélation nulle sinon, comme indiqué en bas à gauche. 

+ Un modèle AR avec phi égal à -0-point-48 a également un décalage 1 autocorrélation -0-point-48. 

+ Cependant, son autocorrélation est positive au décalage 2, négative au troisième décalage et globalement décroissante à zéro en magnitude aux décalages plus élevés, comme indiqué en bas à droite.

####**4. Processus MA et AR : simulations**

+ La simulation de 100 observations de chacun de ces quatre modèles utilisant le même epsilon de bruit blanc fournit une autre comparaison. 

+ Les deux chiffres supérieurs ont un décalage 1 autocorrélation de 0 point 48. 

+ La variation locale est légèrement plus grande pour le modèle AR à droite, mais les deux séries semblent assez similaires. 

+ Les figures du bas ont toutes deux un décalage 1 autocorrélation de -0 point-48. 

+ Le motif oscillant est un peu plus prononcé pour le modèle AR à droite, mais encore une fois les deux séries semblent assez similaires.

####**7. Prévision**

+ Le backtesting ou la mise à jour des prévisions sont des concepts plus avancés qui peuvent être utilisés pour déterminer lequel des deux modèles produit de meilleures prévisions pour un ensemble de données donné. 

+ Une autre approche pour choisir un modèle consiste à utiliser une mesure de qualité d’ajustement telle qu’un critère d’information. 

+ **Plus précisément, les mesures AIC et BIC sont couramment utilisées pour les modèles de séries chronologiques**. 

+ **Le critère d’information est un concept plus avancé, mais pour l’une ou l’autre mesure, une valeur inférieure indique un modèle relativement mieux adapté**.

####**Modèles AR vs AM**


+ ***Comme vous l'avez vu, l'autorégression (AR) et la moyenne mobile simple (MA) sont deux approches utiles pour modéliser des séries chronologiques. Mais comment déterminer si un modèle AR ou MA est plus approprié en pratique ?***

+ **Pour déterminer l'ajustement du modèle, vous pouvez mesurer le critère d'information d'Akaike (AIC) et le critère d'information bayésien (BIC) pour chaque modèle. Bien que les mathématiques sous-jacentes à l'AIC et au BIC sortent du cadre de ce cours, pour vos besoins, l'idée principale est que ces indicateurs pénalisent les modèles avec plus de paramètres estimés, pour éviter le surajustement, et des valeurs plus petites sont préférées. Tous facteurs étant égaux, un modèle qui produit un AIC ou un BIC inférieur à un autre modèle est considéré comme un meilleur ajustement.**

+ **Pour estimer ces indicateurs, vous pouvez utiliser les commandes AIC() et BIC(), qui nécessitent toutes deux un seul argument pour spécifier le modèle en question.**

+ *Dans cet exercice, vous reviendrez aux données du Nil et aux modèles AR et MA que vous avez ajustés à ces données. Ces modèles et leurs prédictions pour les années 1970 (AR_fit) et (MA_fit) sont représentés dans le graphique de droite.*

####**Instructions**

+ Comme première étape de comparaison de ces modèles, utilisez cor() pour mesurer la corrélation entre AR_fit et MA_fit.
+ Utilisez deux appels à AIC() pour calculer l'AIC pour AR et MA, respectivement.
+ Utilisez deux appels à BIC() pour calculer le BIC pour AR et MA, respectivement.

```r
AR <- arima(Nile, order = c(1,0,0))
AR_fit <- Nile - residuals(AR)
```

```r
# Find correlation between AR_fit and MA_fit
cor(AR_fit, MA_fit)
```

```r
# Find AIC of AR
AIC(AR)
```

```r
# Find AIC of MA
AIC(MA)
```

```r
# Find BIC of AR
BIC(AR)
```

```r
# Find BIC of MA
BIC(MA)
```
