###**Pourquoi les gens commercent-ils?**

####**1. Pourquoi les gens commercent-ils?**
+ ***Avant d’introduire quantstrat, établissons les raisons pour lesquelles les gens négocient et ce qu’est un commerce.***

####**2. Qu’est-ce que le trading?**

+ **D’abord et avant tout, une transaction est simplement l’acte d’acheter ou de vendre un actif** 
  - **qu’il s’agisse d’une sécurité financière (comme une action,** 
    + **c’est-à-dire la propriété des capitaux propres d’une entreprise, d’une obligation**, 
    + **c’est-à-dire la propriété de la dette d’une entreprise ou du gouvernement) ou d’un produit physique tangible (comme les produits de base comme l’or, le pétrole, les métaux et le maïs)**. 

+ ***Cela se fait en convertissant les liquidités (telles que les dollars) en propriété de ce produit, et en convertissant le produit (comme les actions d’une société) en espèces (espérons-le pour un profit), ou en prenant livraison d’un bien physique (comme une société pétrolière prenant une cargaison de pétrole brut pour le raffinage).***

####**3. Pourquoi les gens commercent-ils?**
+ Bien que faire un profit en agissant sur les opportunités de trading soit une raison de négocier, il y en a d’autres. 

+ Certaines entreprises dont les activités tournent autour des produits de base entrent sur les marchés financiers pour se protéger de l’impact commercial des mouvements de prix d’un produit sous-jacent (par exemple, Hershey’s pourrait vouloir établir un bon prix pour le chocolat, les compagnies aériennes veulent minimiser l’impact du mouvement des prix du pétrole sur leurs activités, et ainsi de suite. 

+ Par exemple, si le prix du pétrole augmente trop et qu’une compagnie aérienne ne se protège pas contre cela, elle devrait soit augmenter le prix des billets alors que ses concurrents ne le font pas, soit faire beaucoup moins de profit). 

+ Les grandes institutions financières pourraient vouloir augmenter ou diminuer leur exposition à diverses sources de rendement et le risque qui les accompagne. 

+ Au cœur du problème, il y a le fait que les instruments financiers assument le risque, et un gain (prétendument) pour supporter ce risque. 

+ **Le but d’une stratégie de trading systématique est de faire une supposition éclairée quant au moment où le rapport entre la récompense et le risque est suffisamment favorable pour supporter le risque et, à son tour, obtenir une compensation, tout en ne supportant pas ce risque lorsque la rémunération est insuffisante.**

####**4. Types de trading**

+ En termes de mécanique du trading, il existe essentiellement deux types de métiers. 

+ **La divergence, également appelée momentum, ou trading de tendance est la croyance qu’une quantité continuera d’augmenter ou de diminuer si elle a déjà augmenté ou diminué, respectivement.** 

+ **Une classe de fonds spéculatifs connus sous le nom de « conseillers en négociation de matières premières » ou « CTA » ont gagné beaucoup d’argent grâce à des techniques de trading appropriées pour suivre les tendances. Ces catégories de stratégies se caractérisent par de petites pertes sur des marchés sans tendance tout en gagnant beaucoup d’argent lorsqu’une tendance s’établit.**

####**5. Types de trading**
+ **Le style opposé de trading est la convergence, la réversion moyenne, l’oscillation ou le trading de type contrarien.** 

+ **Ce type de philosophie de trading est celui qui tente de prédire quand une certaine quantité va inverser la direction**. 

+ Par exemple, 
  + **la philosophie d’investissement axé sur la valeur du célèbre Warren Buffett consiste à acheter des entreprises après que leur prix a subi une dépréciation considérable, dans l’espoir que cette dépréciation du prix a presque suivi son cours et s’inversera dans un proche avenir.**

  + **Les styles de trading à contre-courant sont classés par de nombreux petits gains tout en subissant des pertes occasionnelles plus importantes lors de l’entrée dans une position de dépréciation trop tôt.**

####**3. Comment ne pas trop s’adapter**

+ **C’est ce qu’on appelle plus communément le surajustement, qui consiste à essayer d’adapter un backtest pour qu’il soit aussi beau que possible (ligne la plus lisse, profit le plus élevé, quels que soient les critères), plutôt que de voir un backtest comme une évaluation d’un système commercial**. 

+ **Afin d’éviter de tels phénomènes, les systèmes de trading devraient s’efforcer d’avoir le moins de paramètres possible** 
  - essentiellement les boutons que vous pouvez tourner pour modifier les paramètres. 
  + Pourquoi? 
    + **Parce que le nombre de configurations système augmente de façon exponentielle avec les paramètres.** 
    + **Autrement dit, considérez une recherche de grille qui teste cinq paramètres distincts pour chaque paramètre**. 
    + **Un système à un paramètre n’aurait besoin que de cinq essais. Un système à deux paramètres nécessiterait vingt-cinq. Un système à trois paramètres aurait déjà besoin de cent vingt-cinq, et au-delà, les chiffres commencent à devenir très élevés très rapidement.**

###**4. Stabilité avec les paramètres du système**

+ De plus, une stratégie de trading ne devrait pas fonctionner très différemment lorsque les paramètres sont légèrement modifiés. 

+ **Par exemple, un système de suivi des tendances doit avoir des résultats similaires pour des paramètres similaires de son filtre de tendance**. 


####**5. Stabilité avec les paramètres du système**

+ ***La bibliothèque TTR de R, écrite par Josh Ulrich, contient une boîte à outils d’indicateurs de trading classiques, tels que la moyenne mobile simple, plus familièrement connue sous le nom de moyenne mobile, et parfois appelée SMA. ***

+ **Dans TTR, la fonction SMA signifie moyenne mobile simple, comme vous l’avez peut-être deviné**. 

+ **Une moyenne mobile simple est un calcul qui prend une moyenne de base sur une période passée, telle que deux cents jours, pour laquelle il y avait des prix**. 

+ **Au fur et à mesure que le temps avance, le dernier prix tombe du calcul, et le plus récent entre, d’où la moyenne des « mouvements »**. 

+ De nombreux conseillers en négoce de matières premières ont beaucoup négocié en utilisant la moyenne mobile à 200 jours. 

+ Certains FNB utilisent encore ce filtre pour activer les positions défensives. 

+ **Bien que la moyenne mobile à 200 jours ne soit pas un chiffre particulièrement magique, c’est une ligne qui affiche où le prix a été au cours des dix derniers mois, car il y a environ deux cents jours ouvrables (c’est-à-dire des jours qui ne sont pas des week-ends ou des jours fériés) dans dix mois.**

####**4. La ligne de tendance**

###**Introduction aux indicateurs**

####**1. Introduction aux indicateurs**
+ ***Bienvenue au troisième chapitre, les indicateurs.***

####**2. Pourquoi utiliser des indicateurs?**

+ ***Pour motiver ce chapitre, le point saillant à garder à l’esprit est que les données de marché sont bruyantes, désordonnées et généralement sujettes à jouer avec les systèmes et les émotions.***

####**3. Que sont les indicateurs?**
+ Afin d’essayer de donner un sens à ces données, les systèmes de négociation utilisent des indicateurs, qui sont des transformations de données qui font une sorte de compromis. 
+ Habituellement, ce compromis est le décalage pour la douceur. 
+ Un indicateur tente de brosser un tableau plus clair de ce qui se passe dans le mouvement des prix de l’actif analysé, mais le paie en ayant besoin d’examiner les données passées, ce qui signifie qu’il pourrait ne pas réagir rapidement à des développements soudains. 
+ Des indicateurs mieux conçus tentent d’obtenir plus de fluidité sans sacrifier la réactivité.

####**4. Exemples d’indicateurs**
+ Sans vertéhé, il existe des indicateurs assez célèbres. 
+ ***La moyenne mobile simple à 200 jours est un filtre, ou plus familièrement, un indicateur de tendance qui est la moyenne de base des deux cents derniers jours, généralement des cours de clôture**. 
+ **Son but est de fonctionner comme une moyenne mobile lente qui signale une attente de hausse des prix lorsqu’une mesure plus rapide est au-dessus, comme les prix de clôture, une moyenne mobile à 50 jours, etc.**

####**5. Exemples d’indicateurs**
+ **Un autre type d’indicateur célèbre est l’indice de force relative, ou RSI, qui est un indicateur oscillant**. 

+ **Le RSI, et d’autres indicateurs oscillants similaires, ne prennent pas de valeurs sur l’échelle du prix de l’actif d’origine, mais prennent plutôt des valeurs comprises entre 0 et 100, 0 à 1 ou des valeurs centrées autour de zéro.** 

+ **Souvent, le but de ces indicateurs oscillants est d’utiliser un bref retour en arrière pour prédire quand un prix a temporairement diminué et devrait augmenter au cours des prochains jours**. 

+ **Entre les systèmes qui peuvent dépendre d’indicateurs d’oscillation à court terme comme un indice de force relative sur 2 jours, et les systèmes de trading à très long terme tels qu’un système croisé à double moyenne mobile, les indicateurs peuvent créer une multitude de stratégies.**

####**6. Dans cette catégorie**

###**Mécanique des indicateurs**

####**1. Mécanique des indicateurs**

+ ***Cette vidéo portera sur la structure de l’ajout d’un indicateur à une stratégie. La fonction qui ajoute un indicateur, sans surprise, est appelée indicateur d’ajout de points. Il comporte cinq parties.***

####**2. Cinq étapes pour appeler des indicateurs**
+ **Pour commencer, on démarre la fonction avec un appel add.indicator**. 

  + **Le premier argument est le nom de votre stratégie**, qui doit être enregistré en tant que variable, car vous l’utiliserez partout. 
    + Jusqu’à présent, nous avons utilisé la **strategy.st** (c’est-à-dire la chaîne de points de stratégie) et continuerons à l’utiliser tout au long de la mise en œuvre de notre stratégie. 
    
  + **Ensuite, vous avez besoin du nom de la fonction que votre indicateur appellera**. 
    + **Les fonctions trouvées dans le paquet TTR, telles que SMA (l’appel de fonction pour le calcul de la moyenne mobile), RSI (l’appel de fonction pour le calcul de l’indice de force relative) et ainsi de suite, sont peu différentes des fonctions R standard, mis à part le fait que puisque tout quantstrat repose sur une base d’un paquet de séries chronologiques, ces fonctions ont produit une sortie ordonnée et utilisent d’autres sous-routines qui dépendent de leurs entrées étant une série chronologique**. 
    
  + En plus du nom de la fonction, il faut paramétrer la fonction dans les arguments entrés dans l’appel **add.indicator**. Il s’agit d’un objet de liste qui sera différent pour chaque fonction. 
  
  + **Enfin, il faut étiqueter la fonction. Cette étiquette est ce que quantstrat utilisera pour trouver votre indicateur lorsque vous écrirez d’autres fonctions qui le rechercheront plus tard.**

####**3. Utilisation de add.indicator()**
+ Voici à quoi ressemble le code. 
+ Essentiellement, il s’agit d’un processus en cinq étapes. 
  + UN : appelez la fonction add.indicator. 
  + Deux : indiquez-lui à quelle stratégie ajouter l’indicateur. 
  + Trois : spécifiez quelle fonction sera utilisée pour générer l’indicateur. 
  + Quatre : ajouter les entrées à la fonction indicateur.
  +  Cinq: ajoutez une étiquette pour nommer la colonne de sortie.

####**4. Une autre façon de penser les indicateurs**
+ Une autre façon de penser à ajouter, ou plutôt à appliquer, des indicateurs est la suivante: 
  + les indicateurs sont similaires à la famille de fonctions d’application de R.

+ **Tout d'abord, la fonction calcule un rapport entre le prix de clôture et la moyenne des prix hauts et bas. Ensuite, il applique un SMA à cette quantité pour lisser le bruit, généralement sur une très courte période, comme deux jours. Enfin, il utilise la fonction runPercentRank() pour prendre un classement en pourcentage courant de ce rapport moyen et le multiplie par 100 pour le convertir en une quantité de 0 à 100.**

+ **Pensez à la façon dont les élèves obtiennent des scores en centile après avoir passé un test standardisé (c'est-à-dire que si un élève obtient un 800 dans sa section de mathématiques, il pourrait être dans le 95e centile à l'échelle nationale). [runPercentRank()](https://www.rdocumentation.org/packages/TTR/versions/0.24.3/topics/runPercentRank) fait la même chose, sauf dans le temps. Cet indicateur fournit le classement de la dernière observation lorsqu'elle est prise dans le contexte sur une période passée spécifiée par l'utilisateur. Par exemple, si quelque chose a une valeur runPercentRank de 0,90 lors de l'utilisation d'une période d'analyse de 126, cela signifie qu'il se situe dans le 90e centile par rapport à lui-même et aux 125 dernières observations.**

+ *Votre travail consiste à mettre en œuvre cet indicateur et à l'enregistrer en tant que DVO. Une partie du code nécessaire a été fournie et les packages quantstrat, TTR et quantmod sont chargés dans votre espace de travail.*

####**Instructions**

+ Créez et nommez une fonction, DVO, pour l'indicateur décrit ci-dessus. Les trois arguments de votre fonction seront HLC, navg (par défaut à 2) et percentlookback (par défaut à 126).
+ Le rapport de la clôture (Cl()) de HLC divisé par la moyenne des prix haut (Hi()) et bas (Lo()) est calculé pour vous.
+ Utilisez SMA() pour implémenter une moyenne mobile de ce rapport, paramétrée par l'argument navg. Enregistrez cela comme avgratio.
+ Utilisez runPercentRank() pour implémenter un système de classement en pourcentage pour l'avgratio.

```r
# Delare the DVO function
DVO <- function(HLC, navg = 2, percentlookback = 126) {
  
  # Compute the ratio between closing prices to the average of high and low
  ratio <- Cl(HLC)/((Hi(HLC) + Lo(HLC))/2)
  
  # Smooth out the ratio outputs using a moving average
  avgratio <- SMA(ratio, n = navg)
  
  # Convert ratio into a 0-100 value using runPercentRank()
  out <- runPercentRank(avgratio, n = percentlookback, exact.multiplier = 1) * 100
  colnames(out) <- "DVO"
  return(out)
}
```

###**sigFormula**
####**1. sigFormula**
####**2. À propos de sigFormula**

+ La dernière fonction de signal à introduire est sigFormula. 

+ Simplement, il s’agit d’une fonction qui utilise l’évaluation des cordes pour mélanger et faire correspondre les indicateurs et les signaux existants afin de former des signaux d’ensemble. 

+ Par exemple, dans de nombreux systèmes de négociation, un signal (tel qu’un oscillateur signalant que le prix a connu un repli suffisant) ne sera pris en compte que s’il se produit dans un environnement de marché favorable (comme une moyenne mobile simple à 50 jours supérieure à une moyenne mobile simple de 200 jours). 

+ Par exemple, dans un marché où les prix ont chuté, comme le marché boursier américain en 2008, les prix ne connaissaient pas simplement une baisse temporaire, mais étaient en train de baisser de 55 %. 

+ Pour ceux qui ne le savent pas, un drawdown est le plus grand pic à creux que l’on ait subi. 

+ Par exemple, une stratégie qui commence avec cent dollars, rapporte cinquante dollars, puis redescend à soixante-quinze dollars aurait un tirage maximum de cinquante pour cent. 

+ Dans un repli aussi massif du marché, lorsqu’une tendance à long terme est fortement négative, l’adoption d’une position à contre-courant pourrait entraîner des pertes importantes. 

+ En bref, un commerce à contre-courant devrait s’assurer qu’il achète un repli temporaire, et non une baisse importante.

####**3. Structure**
+ À cette fin, la fonction sigFormula permet une flexibilité exceptionnelle dans la création de signaux qui sont des combinaisons d’autres signaux. 

+ Pour utiliser une fonction sigFormula, vous pouvez l’appeler de la même manière que les fonctions sigComparison, sigCrossover et sigThreshold. 

+ Cependant, plutôt que d’avoir les arguments explicitement énoncés comme arguments à la fonction de signal, on passe dans une chaîne logique à l’argument de formule de sigFormula. 

+ Si cela semble intimidant, ce n’est pas le cas. 

+ L’instruction passée à l’argument de formule de sigFormula n’est rien de plus que le type d’argument que l’on écrirait à l’intérieur d’une instruction if. 

+ Par exemple, si l’énoncé 1 et l’énoncé 2 doivent être vrais, alors en base R, on écrirait if (statement_1 & statement_2). 

+ Ce n’est pas différent pour l’argument de la formule, mis à part le fait que le texte est une citation et n’a pas besoin de l’instruction if et des parenthèses. Personnellement, je l’utilise pour combiner des signaux, plutôt que des indicateurs.

###**Introduction aux règles**

####**1. Introduction aux règles**
+ ***Bienvenue dans le chapitre sur les règles dans quantstrat.***

####**2. Que sont les règles?**


+ ***Les règles sont un moyen pour vous de spécifier exactement comment vous allez créer votre transaction une fois que vous décidez que vous souhaitez exécuter un signal. Autrement dit, étant donné qu’une valeur de signal particulière a généré une déclaration VRAIE, combien cherchez-vous à acheter ou à vendre? En termes de complexité, les règles sont les plus impliquées des trois structures de quantstrat - indicateurs, signaux et règles. Alors que les indicateurs et les signaux sont assez simples, les règles ont beaucoup plus de parties et beaucoup de personnalisation. Bien que la plupart de cette personnalisation dépasse le cadre de ce cours, vous saurez éventuellement comment utiliser des règles en conjonction avec des fonctions de dimensionnement des ordres et pourrez utiliser des ordres au marché qui peuvent modifier la taille de la position en fonction de diverses autres entrées.***

####**3. Utilisation de add.rule()**
+ En raison de la structure compliquée de add-dot-rule(), vous serez exposé à de nouveaux arguments dans chaque vidéo de ce chapitre. 
+ Tout d’abord, il existe deux types de règles dans le champ d’application de cette classe: les règles d’entrée et les règles de sortie. 

+ Les règles d’entrée sont des règles qui visent à établir une position, c’est-à-dire, dans ce cas, l’achat d’actions d’un FNB. 

+ Les règles de sortie sont des règles dédiées à retirer votre argent d’un actif et à le reconvertir en espèces, prêt à être réutilisé. 

+ Dans le cadre d’un appel add-dot-rule, l’entrée de type de règle prend la valeur enter ou exit. 

+ L’une des principales différences entre les règles et les indicateurs ou signaux est que les règles n’utilisent qu’une seule fonction fourre-tout pour presque toutes les entrées. Il s’appelle ruleSignal. 

+ La plupart des arguments pour les règles serviront d’arguments à ruleSignal. 

+ En outre, comme les indicateurs et les signaux qui les précèdent, les règles font également référence aux colonnes existantes de votre stratégie. 

+ Ces colonnes existantes sont désignées par l’argument label dans vos appels add-dot-signal. 

+ L’argument sigcol est l’entrée que quantstrat utilise pour trouver les colonnes de signal existantes dans votre stratégie. 

+ Ensuite, ruleSignal prend un argument sigval. 

+ Cet argument indique à quantstrat quelle valeur dans la colonne de signal particulière déclenche une transaction. 

+ Bien que cette valeur puisse être autre que TRUE, il est assez recommandé de définir cet argument sur TRUE si vos colonnes de signal sont configurées correctement pour ne renvoyer TRUE que lorsque vous souhaitez potentiellement conclure une transaction.

####**1. Fonctions de dimensionnement des commandes**
+ ***Enfin, un dernier aspect important de l’utilisation des règles est la possibilité d’utiliser les fonctions de dimensionnement des commandes.***

####**2. Que sont les règles?**
+ **Les fonctions de dimensionnement des ordres sont, comme vous pouvez l’imaginer, des fonctions qui déterminent la quantité d’un actif que vous achetez ou vendez, si vous n’utilisez pas l’argument orderqty. Les fonctions de dimensionnement de commande créent des tailles de position variables, basées sur les entrées de la fonction de dimensionnement de commande.**

####**3. Que sont les règles?**
+ ***Bien que l’écriture de fonctions de dimensionnement d’ordre implique l’introduction d’une syntaxe complexe et dépasse le cadre de cette classe, la possibilité de les appeler est assez simple. Lors de l’utilisation d’une fonction de dimensionnement d’ordre, l’argument orderqty de ruleSignal devient effectivement obsolète- ou n’est plus significatif pour les arguments de la fonction, et remplacé par la sortie de la fonction de dimensionnement d’ordre.***

####**4. Que sont les règles?**
+ Pour utiliser une fonction de dimensionnement de commande, vous passez la fonction de dimensionnement de l’ordre en tant qu’objet à l’argument osFUN. 

+ En outre, l’argument de la fonction de dimensionnement de l’ordre s’insère dans la même liste d’arguments que le reste des entrées pour ruleSignal. 

+ Pour votre stratégie, vous utiliserez la fonction de dimensionnement des commandes appelée osMaxDollar. 

+ Cette fonction obtient simplement une position égale à tradeTaille de l’actif, arrondie à l’unité la plus proche de l’actif. 

+ Donc, si votre tradeSize est de cent cinq dollars et que le prix est de vingt dollars, vous obtiendrez cinq unités (plus familièrement appelées actions) de l’actif. 

+ De même, le maxSize est un argument pour le montant maximum de dollars que votre position peut être. 

+ Habituellement, ce montant est égal à la taille de la transaction, ce qui signifie qu’une fois que vous avez une position, elle ne changera probablement pas de taille. 

+ Essentiellement, on peut penser à utiliser une fonction de dimensionnement d’ordre comme une sorte d’instruction apply imbriquée. 

+ Alors qu’une instruction apply dans R prend une fonction comme argument, un ruleSignal qui prend une fonction de dimensionnement d’ordre est similaire à la transmission d’une fonction à une instruction apply qui a besoin d’une autre fonction en entrée. 

+ Essentiellement, les fonctions de dimensionnement d’ordre nécessitent un niveau supplémentaire d’encapsulation de fonction, tandis que les arguments de ces fonctions de dimensionnement d’ordre partagent de l’espace avec les arguments de ruleSignal.

####**2. Le graphique. Fonction Posn**
+ La fonction d’éclairage de Quantstrat consiste à examiner vos positions au fil du temps.
+  Pour ce faire, utilisez la fonction chart-dot-Posn. Jetons un coup d’œil à la fonction chart-dot-Posn pour notre stratégie.

####**3. À quoi ça ressemble**
+ L’utilisation de ce code fait apparaître un graphique du prix, ainsi que les positions que vous avez prises au fil du temps. 
+ En outre, il affichera la taille de vos positions au cours de la stratégie, qui sont les rectangles bleu et jaune, ainsi qu’une courbe de trésorerie ou de profits et pertes basée sur les actions. 
+ Il s’agit de la ligne verte, et il y a un graphique de tirage en dessous, qui est une ligne rouge. 
+ Pour préciser, chaque rectangle bleu est une transaction. 
+ Lorsque sa taille augmente 
  - c’est-à-dire que la quantité bleue surmontée d’un point jaune s’éloigne de zéro, cela signifie que la stratégie a pris position 
  - en d’autres termes, a fait un pari. 
  
+ Une position supérieure à zéro est un pari sur l’augmentation des prix, et une position inférieure à zéro est un pari sur la baisse des prix. 
+ Comme vous pouvez le voir dans cette stratégie, nous n’avons jamais une position inférieure à zéro. 
+ En passant à la ligne verte, qui est la courbe des capitaux propres basée sur les profits et pertes, cette ligne verte vous montre la valeur globale du compte au fil du temps. 
+ Plus la ligne monte, meilleure est la stratégie. Idéalement, cette ligne verte devrait continuer à monter et à droite, ce qui signifierait un système toujours rentable. 
+ Contrairement à la courbe des actions, la ligne rouge est le graphique de drawdown. Pour revoir, le drawdown est le montant d’argent perdu depuis que le portefeuille a atteint un nouveau sommet. 
+ Dans quantstrat, la fonction chart-dot-Posn inclura un graphique de drawdown basé sur le profit et la perte juste sous la courbe des actions, dont nous venons de parler. 
+ Contrairement à la courbe des actions, que tout le monde veut voir s’écarter de zéro, vous voulez voir le graphique de baisse avoir le moins d’action possible. 
+ C’est-à-dire que la ligne rouge ne peut rester qu’en dessous de zéro, et vous voulez qu’elle reste aussi proche de zéro que possible aussi souvent que possible. 
+ Lorsque le graphique de tirage s’éloigne de zéro, cela signifie toujours qu’une stratégie perd de l’argent.

####**3. Rapport de Sharpe**
+ L’obtention de ces données vous permet de prendre un ratio de Sharpe de vos profits et pertes. Le ratio de Sharpe est le rapport entre le montant d’argent que vous avez gagné et la volatilité ou le risque du processus. 

+ Simplement, il s’agit d’un rapport entre la récompense et le risque. 

+ Syntaxiquement, l’argument géométrique doit être défini sur FAUX car il ne s’agit pas de données basées sur les retours, et dans quantstrat, l’équité du compte ne se compose pas continuellement à chaque transaction. 

+ Pour le dire en termes plus simples, dans quantstrat, alors que vous pouvez faire un profit sur une transaction, quantstrat ne le met pas à jour à chaque nouvelle transaction. 

+ Au lieu de cela, une meilleure façon de penser à une simulation quantstrat est que les profits sont retirés du système. 

+ Cela signifie que le deuxième trade ne dépend pas du résultat du premier trade, et lorsque vous calculez les résultats de la stratégie, vous devez constamment garder cela à l’esprit.

####**4. Obtenir des retours**
+ De plus, vous pouvez obtenir tous les rendements d’instruments pour votre stratégie. 
+ Les rendements dans une simulation quantstrat sont définis comme le rapport entre le profit ou la perte sur une transaction donnée, divisé par les capitaux propres initiaux définis au début de la simulation. 
+ La commande PortfReturns crée une sortie qui contient les rendements de tous les instruments individuels (dans cet exemple, LQD et dans les exercices, il s’agit de SPY) au cours de la simulation. 
+ Comme vous n’avez exécuté la simulation que sur un seul instrument, vous n’obtiendrez qu’un seul vecteur en retour.

####**5. Obtenir le ratio de Sharpe pour les retours**
+ Une fois que vous avez vos retours, vous pouvez utiliser de nombreuses fonctions du package PerformanceAnalytics pour effectuer des analyses supplémentaires. 

+ Dans ce cas, vous pouvez obtenir le ratio de Sharpe à partir des retours. Dans ce cas, définissez géométrique sur FAUX, car les rendements dans quantstrat sont basés sur les profits et pertes divisés par les capitaux propres initiaux de votre compte.
