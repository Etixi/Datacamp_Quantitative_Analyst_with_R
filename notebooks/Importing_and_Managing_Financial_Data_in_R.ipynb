###**Introduction à Quandl**


####**1. Introduction à Quandl**

+ ***Maintenant que vous avez appris les bases de getSymbols(), j’aimerais vous présenter une autre méthode pour importer des données dans votre session R.***

####**2. Qu’est-ce que Quandl?**

+ ***Quandl est un service de données qui donne accès à de nombreux ensembles de données financières et économiques de divers fournisseurs.*** 

+ De nombreux ensembles de données sont accessibles gratuitement, mais certains nécessitent un abonnement payant. 

+ **Quandl fournit une interface de programmation d’application, ou API, qui vous permet d’importer des données à partir d’une grande variété de langages, y compris Excel, Ruby, Python et, bien sûr, R**. 

+ **Le package R est également appelé Quandl et contient une fonction que vous apprendrez à utiliser pour importer des données à partir des bases de données Quandl. Le nom de la fonction est, vous l’avez deviné, Quandl().**

####**3. Quandl() contre getSymbols()**

+ Comparons la fonction Quandl() avec getSymbols(). 

+ Les deux fonctions permettent d’accéder à plusieurs sources de données. 

+ **Bien que getSymbols() puisse créer un objet pour vous, la fonction Quandl() retourne toujours les données qu’elle importe, vous devez donc affecter les données à un objet**. 

+ **Et rappelez-vous qu’avec getSymbols(), vous utilisez les arguments Symbols et src pour spécifier l’instrument et la source de données. Avec Quandl(), l’instrument et la source de données sont spécifiés par un seul argument nommé code. Notez que la source de données est la première, puis une barre oblique, puis l’instrument.**

####**4. Quandl() contre getSymbols()**

+ **Comme getSymbols(), la fonction Quandl() peut importer les données dans plusieurs classes différentes d’objets, y compris un bloc de données, un objet xts ou zoo, ou un objet ts (du package stats) ou un objet timeSeries du package timeSeries de la suite RMetrics.** 

+ **Vous pouvez contrôler la classe de l’objet qu’il renvoie en modifiant l’argument type. Une autre différence entre getSymbols() et Quandl() est que getSymbols() retourne un objet xts par défaut, tandis que la fonction Quandl() retourne une trame de données par défaut.**

####**5. Entraînons-nous!**
+ Maintenant, c’est à votre tour de vous entraîner en utilisant la fonction Quandl().

####**Présentation de Quandl()**

+ ***Semblable à la façon dont le package quantmod fournit getSymbols() pour importer des données à partir de diverses sources, le package Quandl permet d'accéder aux bases de données Quandl via une fonction simple : Quandl().**

+ **Rappelez-vous que getSymbols() utilise les arguments Symbols et src pour spécifier respectivement l'instrument et la source de données. La fonction Quandl() spécifie à la fois la source de données et l'instrument via son argument de code, sous la forme "DATABASE/DATASET".**

+ Quandl() diffère de getSymbols() de deux autres manières :

  + **Quandl() renvoie un data.frame par défaut**.
  + **Quandl() n'affectera pas automatiquement les données à un objet.**


+ *Si vous prévoyez d'importer beaucoup de données à l'aide de Quandl(), vous pouvez envisager d'ouvrir un compte gratuit avec eux afin d'obtenir une clé API.*

###**Définition des arguments par défaut pour getSymbols()**

####**1. Définition des arguments par défaut pour getSymbols()**

+ ***Vous avez appris comment getSymbols() permet d’accéder à plusieurs sources de données et que getSymbols() importe les données de Yahoo Finance par défaut. Que se passe-t-il si votre analyse dépend de données provenant principalement d’une source autre que Yahoo Finance ? Devez-vous spécifier « src » pour chaque appel getSymbols() ? Dans cette vidéo, vous allez apprendre à utiliser la fonction setDefaults() pour définir une nouvelle source de données par défaut pour getSymbols() et comment utiliser setDefaults() pour personnaliser les arguments par défaut transmis aux méthodes getSymbols() sous-jacentes.***

####**2. getSymbols() « méthodes »**

+ **Que sont les méthodes getSymbols() ?** 

+ **La fonction getSymbols() ne contient pas réellement de code pour importer des données**. 

+ **Au lieu de cela, il détermine la source de données pour chaque symbole, puis appelle la « méthode » getSymbols() pour effectuer l’importation réelle.**

+ **Par exemple, lorsque vous importez des données GDP à partir de FRED, vous appelez getSymbols() avec « src » défini sur « FRED ».** 

+ **Ensuite, getSymbols() appelle la fonction getSymbols.FRED(). Notez que vous ne devez pas appeler ces « méthodes » directement.**

####**3. Utilisez setDefaults() pour modifier la source de données par défaut**

+ Regardons un exemple d’utilisation de setDefaults(). 
+ Supposons que vous souhaitiez remplacer la source de données par défaut de Yahoo Finance par FRED. 
+ **Le premier argument de setDefaults() est la fonction getSymbols()**. 
+ Les autres arguments de setDefaults() sont des paires de noms d’arguments et de nouvelles valeurs par défaut. Dans ce cas, il vous suffit de définir « src » sur « FRED ». Voilà! Maintenant, si vous appelez getSymbols() sans argument « src », il importera les données de FRED.

####**4. setDefaults()**
+ Quelques points importants à garder à l’esprit à propos de setDefaults(). 

+ Vous devez spécifier des arguments car nom est égal à paires de valeurs.

####**1. Manipulation des symboles d’instruments qui entrent en conflit ou qui ne sont pas des noms R valides**

####**2. Noms syntaxiquement valides**
+ Les noms syntaxiquement valides peuvent contenir des lettres, des chiffres, un point (ou un point) et un trait de soulignement. 
+ Ils doivent commencer par une lettre ou un point suivi d’un non-chiffre. 
+ Il est en fait possible de créer des objets R avec des noms syntaxiquement invalides, mais ils doivent être accessibles de manière non conventionnelle.

####**3. Accès aux objets avec des noms non syntaxiques (1)**
+ Tous les symboles boursiers ne sont pas des noms syntaxiquement valides. 
+ Par exemple, la plupart des symboles d’indice sur Yahoo Finance commencent par un circonflexe, comme l’indice S & P 500. 
+ Étant donné que les objets R avec des noms syntaxiquement invalides peuvent être difficiles à utiliser, getSymbols() supprime automatiquement l’accent circonflexe de l’objet qu’il crée et des noms de colonne de l’objet.

####**4. Accès aux objets avec des noms non syntaxiques (2)**
+ Mais getSymbols() ne convertit pas toujours le symbole boursier et les noms de colonne en noms syntaxiquement valides, comme vous pouvez le voir pour l’indice composite de la Bourse de Shanghai. 
+ Vous obtiendrez une erreur si vous essayez d’utiliser cet objet comme s’il avait un nom valide.

####**5. Accès aux objets avec des noms non syntaxiques (3)**
+ Si getSymbols() charge des données dans votre espace de travail et que l’objet n’a pas de nom syntaxiquement valide, les deux façons les plus simples d’accéder à l’objet renvoyé sont d’entourer le nom de l’objet de coches rétroactives ou d’utiliser la fonction get().

####**6. Nom valide pour un instrument**
+ Si vous savez que le symbole boursier n’est pas un nom syntaxiquement valide, vous pouvez définir auto-dot-assign = FALSE dans votre appel getSymbols() afin de pouvoir affecter directement l’objet à un nom syntaxiquement valide. 
+ Dans ce cas, les noms de colonne ne seront probablement pas non plus syntaxiquement valides. Cela peut poser problème si vous devez utiliser une fonction qui s’attend à ce que les noms de colonne soient des noms R valides. 
+ Vous pouvez utiliser paste() pour créer des noms de colonne syntaxiquement valides, puis utiliser la fonction d’affectation colnames() pour les affecter à l’objet.

####**7. Nom valide pour plusieurs instruments**

+ Vous pouvez également utiliser setSymbolLookup() pour mapper le symbole boursier d’une source de données à un nom d’objet R syntaxiquement valide.

+ **La création d’un jeu de données régulier à partir des données irrégulières brutes peut faciliter la fusion avec d’autres séries, ce qui facilite l’analyse**.

+ **Les objets xts de largeur nulle sont utiles lorsque vous souhaitez ajouter ou supprimer des lignes d’un autre objet xts, sans ajouter de colonne.**

####**5. Création de données régulières à partir de données irrégulières**
+ Lorsque vous créez des données régulières à partir de données irrégulières, vous devez généralement ajouter des observations à chaque date-heure de la séquence normale, car il est très peu probable que des données irrégulières se produisent exactement à une date-heure normale. 

+ Ces observations ajoutées seront NA dans le résultat, car elles ne se produisent pas dans les données irrégulières et l’objet xts de largeur nulle n’a pas de colonne de données.

###**Agrégation à basse fréquence**

####**1. Agrégation à basse fréquence**
+ ***Bien qu’il soit courant de faire des analyses sur les données financières quotidiennes, la plupart des données économiques ne sont publiées que mensuellement ou trimestriellement.***

####**2. Données basse fréquence**
+ Bien qu’elles ne se produisent que mensuellement ou trimestriellement, les observations comprennent généralement un horodatage avec un jour du mois. 

+ Certains fournisseurs de données utilisent le premier jour de la période d’observation, tandis que d’autres utilisent le dernier jour. 

+ Et il n’est pas rare d’utiliser le mois du milieu pour les données trimestrielles.

####**3. Exemple**
+ Regardons un exemple d’essayer de fusionner deux ensembles de données basse fréquence avec des conventions d’horodatage différentes. 

+ Supposons que vous souhaitiez comparer le taux quotidien à échéance constante des bons du Trésor à 10 ans avec le produit intérieur brut des États-Unis, qui n’est publié que trimestriellement. 

+ Les symboles FRED pour ces séries sont respectivement DGS10 et PIB.

####**4. Fusionner les données agrégées avec les données à basse fréquence**
+ **Vous pouvez utiliser apply-dot-quarterly() pour calculer la médiane des valeurs quotidiennes pour chaque trimestre.** 

+ **Rappelez-vous que la sortie apply-dot-quarterly() utilisera l’horodatage de la dernière observation du trimestre. Et les données du PIB sont horodatées avec le premier jour du trimestre, de sorte que les colonnes des données fusionnées n’ont pas d’horodatage en commun.**

####**5. Classes date-heure basse fréquence**
+ Le forfait zoo fournit deux classes pour gérer cette situation. 

+ **La classe yearmon représente les observations mensuelles et la classe yearqtr représente les observations trimestrielles.** 

+ **Ces classes éliminent l’ambiguïté du jour qui représente un mois ou un trimestre spécifique.**

####**6. Convertir l’index en fréquence la plus basse**
+ Revenons à notre exemple du PIB et des taux du Trésor. + Il est souvent plus facile d’utiliser la classe avec la fréquence la plus basse compte tenu de vos données. 
+ **Dans ce cas, vous pouvez convertir les deux index en classe yearqtr trimestrielle à l’aide de as-dot-yearqtr()**. 
+ **Affectez ensuite la sortie de as-dot-yearqtr() à l’index de chaque objet**. 

+ **Maintenant, la fusion des deux objets fonctionne et il n’y a pas de valeurs manquantes**. 

+ **Si vous ne pouvez pas utiliser la classe yearqtr pour une raison quelconque, la deuxième meilleure option consiste à vous aligner sur le premier horodatage de la période, car les mois et les trimestres ont un nombre de jours différent.**

####**7. Aligner sur l’horodatage de début de période**
+ **Vous pouvez le faire en utilisant na-dot-locf() sur le résultat fusionné, avec l’argument fromLast défini sur TRUE afin que la dernière observation soit reportée en arrière**. 

+ Maintenant, les valeurs manquantes sont remplies, de sorte que le taux d’intérêt médian a maintenant une valeur non-NA pour la première observation de chaque trimestre. 

+ Mais il y a encore deux observations par trimestre; un au début et un à la fin. 

+ Vous pouvez extraire uniquement la première observation en sous-définissant l’index de l’objet avec les valeurs d’index que vous souhaitez conserver. 

+ Dans ce cas, c’est l’indice du PIB.

####**11. Exemple de fractionnement d’actions**
+ Alors, que sont les fractionnements et les dividendes? Un fractionnement d’actions est lorsqu’une société augmente simultanément le nombre d’actions en circulation et diminue le cours de l’action. 

+ Cela signifie que la valeur globale de l’entreprise ne change pas. 

+ Par exemple, un fractionnement d’actions 2 pour 1 donnerait aux investisseurs 2 actions pour 1 action qu’ils possèdent, mais réduirait le cours de l’action de 50% en même temps.

####**12. Exemple de dividende en actions**
+ Un dividende est lorsqu’une entreprise décide de restituer du capital aux investisseurs, souvent payé en espèces. 

+ Contrairement aux fractionnements, les dividendes réduisent la valeur de l’entreprise parce que l’argent quitte réellement l’entreprise. 

+ Cela devrait théoriquement entraîner une baisse du prix du montant du dividende. 

+ Même si le cours de l’action de la société baisse, le rendement de l’investisseur n’est pas affecté, car il a reçu le paiement du dividende compensatoire.

####**Ajuster les données univariées pour les fractionnements et les dividendes**
+ ***Si vous n'avez que des prix de clôture, vous pouvez les ajuster avec adjRatios(). Il a 3 arguments : scissions, dividendes et clôture. Il renvoie un objet xts avec des ratios d'ajustement de fractionnement et de dividende dans les colonnes "Split" et "Div", respectivement.***

+ **Vous devez fournir des données fractionnées via l'argument fractionnements pour calculer le rapport de fractionnement. Pour calculer le ratio de dividende, vous devez fournir des dividendes bruts et des prix bruts via les dividendes et les arguments proches, respectivement.**

+ **Une fois que vous avez les ratios d'ajustement de fractionnement et de dividende, vous calculez le prix ajusté en multipliant le prix non ajusté par les ratios d'ajustement de fractionnement et de dividende.**

*Votre espace de travail contient les fractionnements, les dividendes et les raw_dividends des exercices précédents, ainsi que les données AAPL de Yahoo Finance.*

####**Instructions**

+ Utilisez adjRatios() pour calculer les ratios d'ajustement de fractionnement et de dividende. Attribuez la sortie à des rapports.
+ Calculez la clôture ajustée pour AAPL en extrayant le prix de clôture et en le multipliant par les deux ratios d'ajustement et affectez-le à aapl_adjusted.
+ Regardez les premières lignes de la colonne de clôture ajustée fournie par Yahoo Finance.
+ Regardez les premières lignes du prix de clôture ajusté que vous avez calculé et comparez-le à la clôture ajustée de Yahoo Finance.

```r
# Calculate split and dividend adjustment ratios
ratios <- adjRatios(splits = splits, dividends = raw_dividends, close = Cl(AAPL))

# Use the Split and Div columns to calculate adjusted close for AAPL
aapl_adjusted <- Cl(AAPL) * ratios[, "Split"] * ratios[, "Div"]

# Look at first few rows of adjusted close for AAPL
head(Ad(AAPL))
```
