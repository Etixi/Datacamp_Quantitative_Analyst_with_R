+ Il est temps d'appliquer ! Plus tôt, Lore vous a parlé des rendements financiers.

+ Supposons que vous ayez 100 $. En janvier, vous faites un retour de 5% sur cet argent. 
+ Tu as combien fin janvier ? 
+ Eh bien, vous avez 100 % de votre argent de départ, plus 5 % supplémentaires : 
  + 100 % + 5 % = 105 %.

  + En décimal, c'est 1 + 0,05 = 1,05. 

  + Ce 1,05 est le multiplicateur de rendement pour janvier, et vous multipliez vos 100 $ d'origine par celui-ci pour obtenir le montant que vous avez à la fin de janvier.

  + 105 = 100 * 1,05

+ Soit en termes de variables :

  + post_jan_cash <- starting_cash * jan_mult

+ Un moyen rapide d'obtenir le multiplicateur est :

  + multiplier = 1 + (return / 100)

####**Instructions**
+ Votre nouvelle trésorerie de départ, le rendement de janvier et le multiplicateur de rendement de janvier ont été définis pour vous.

```r
# Variables for starting_cash and 5% return during January
starting_cash <- 200
jan_ret <- 5
jan_mult <- 1 + (jan_ret / 100)

# How much money do you have at the end of January?
post_jan_cash <- starting_cash * jan_mult

# Print post_jan_cash
post_jan_cash

# January 10% return multiplier
jan_ret_10 <- 10
jan_mult_10 <- 1 + (jan_ret_10 / 100)

# How much money do you have at the end of January now?
post_jan_cash_10 <- starting_cash * jan_mult_10

# Print post_jan_cash_10
post_jan_cash_10
```

####**Rendements financiers (2)**
+ Faisons un peu plus d'argent. 
+ Si, en février, vous gagnez encore 2 % sur votre argent, comment calculeriez-vous le montant total à la fin février ? 
+ Vous savez déjà que le montant fin janvier est de 100 $ * 1,05 = 105 $. 
+ Pour passer de fin janvier à fin février, il suffit d'utiliser un autre multiplicateur !

  + 105 $ * 1,02 = 107,1 $

+ Ce qui équivaut à :

  + 100 $ * 1,05 * 1,02 = 107,1 $

+ *Dans ce dernier formulaire, vous voyez l'effet des deux multiplicateurs sur vos 100 $ d'origine. En fait, ce formulaire peut vous aider à trouver le rendement total sur les deux mois. La bonne façon de procéder consiste à multiplier les deux multiplicateurs ensemble : 1,05 * 1,02 = 1,071. Cela signifie que vous avez gagné 7,1 % au total sur la période de 2 mois.*

####**Instructions**
+ Votre trésorerie de départ et les rendements de janvier et février ont été donnés.
+ Utilisez-les pour calculer les multiplicateurs de rendement de janvier et février : jan_mult et feb_mult.
+ Utilisez ces multiplicateurs et starting_cash pour trouver votre total_cash à la fin des deux mois.
+ Imprimez total_cash pour voir comment votre argent a augmenté !

```r
# Starting cash and returns 
starting_cash <- 200
jan_ret <- 4
feb_ret <- 5

# Multipliers
jan_mult <- 1 + (jan_ret / 100)
feb_mult <- 1 + (feb_ret / 100)

# Total cash at the end of the two months
total_cash <- starting_cash * jan_mult * feb_mult

# Print total_cash
total_cash
```

####**2. Numérique**
+ Les données numériques sont des nombres décimaux comme 42 points 5. 
+ Les données numériques peuvent être les cours des actions, les flux de trésorerie ou les revenus. 
+ Un type particulier de données numériques est constitué d’entiers. 
+ Les entiers sont des nombres entiers qui n’ont pas de décimale. 
+ Par défaut, les nombres entiers comme 5 sont stockés en tant que données numériques, sauf si vous indiquez spécifiquement à R que vous souhaitez qu’il stocke 5 en tant qu’entier. Pour ce faire, ajoutez un L majuscule après le nombre. 
+ Vous utiliserez presque toujours des chiffres sur des entiers dans R, donc ce comportement par défaut n’a rien à craindre.

####**3. Caractère**
+ Les caractères, également appelés chaînes, sont des valeurs de texte telles que « Hello world » ou « forty ». 
+ Les données de caractère peuvent être une séquence de noms, ou des catégories pour vos données, ou même une cote de crédit pour les obligations. 
+ Notez les guillemets indiquant à R que vous lui donnez des données de caractère. 
+ Les nombres pairs peuvent être stockés en tant que données de caractère si vous les mettez entre guillemets. 

####**4. Logique**
+ Et enfin, les logiques sont les valeurs dites booléennes de TRUE et FALSE. 
+ R exige que true et false soient en majuscules, ce qui provoquerait une erreur. 
+ Techniquement, NA, pour désigner une valeur manquante, est également logique, mais nous n’entrerons pas dans les détails ici.

####**5. Variables et types de données**
+ Jusqu’à présent, vous avez stocké des données numériques dans des variables, mais vous auriez pu tout aussi facilement stocker des données logiques ou de caractères dans une variable.

+ Par défaut, il remplira d’abord _down_ avant de passer à la colonne suivante.

+ Une chose à noter à propos des matrices est que, comme les vecteurs, vous ne pouvez y mettre qu’un seul type de données.

###**Qu’est-ce qu’une trame de données ?**

####**1. Qu’est-ce qu’une trame de données ?**
+ Il est temps d’en apprendre davantage sur l’une des structures de données les plus courantes et les plus utiles de R. Bloc de données.

####**2. Cadre de données**

+ Un bloc de données est utilisé pour stocker une table de données et n’est rien de plus qu’une grille avec des lignes et des colonnes, et vous les avez probablement déjà utilisés dans d’autres programmes. 

+ Vous pourriez dire, attendez, n’est-ce pas juste une matrice? 

+ Presque! Bien qu’une trame de données ait une apparence similaire à celle d’une matrice, en ce sens qu’elle comporte des lignes et des colonnes, elle diffère en ce sens que vous pouvez avoir plusieurs types de données dans une seule trame de données. 

+ Cela signifie que vous pouvez mettre un vecteur de caractère dans une colonne et un vecteur numérique dans une autre, comme ce qui se fait ici.

####**3. Cadres de données et amis**

+ Pour voir cela en action, revenons à l’exemple de Dan et Rob qui vous doivent de l’argent. 

+ Au lieu de lier ensemble les vecteurs comme nous l’avons fait avec les matrices, nous pourrions créer une trame de données pour contenir cette information. 

+ Cela ressemblerait à ceci. 

+ Remarquez qu’il y a une colonne de caractère, nom, pour représenter qui vous doit l’argent, et une colonne numérique, paiement, pour représenter combien ils vous doivent. 

+ Chaque colonne représente une variable unique et chaque ligne représente une observation de données. 

+ Ce type de structure de variables en colonnes et d’observations en lignes est extrêmement puissant et vous facilitera infiniment la vie lorsque vous essayez de sous-ensemble, de tracer ou de résumer vos données plus tard. 

+ La création du bloc de données est très similaire aux autres structures de données que vous avez créées jusqu’à présent. 

+ Il vous suffit de passer chaque vecteur que vous souhaitez ajouter au bloc de données. 

+ Rappelez-vous simplement que chaque vecteur doit avoir la même longueur!

####**4. Nommez ce cadre!**
+ Si vous souhaitez modifier les noms de vos colonnes, vous pouvez le faire à l’aide de la fonction colnames. 

+ Par exemple, si vous souhaitez renommer la colonne « nom » en « ami » et la colonne « paiement » en « argent », vous pouvez voir que le processus est très similaire à la dénomination des vecteurs. 

+ Le vecteur de caractère « ami » et « argent » est attribué aux colnoms de la dette. 

+ En outre, comme les vecteurs, cela aurait pu être spécifié lors de la création de la trame de données.
