####**2. Données de séries chronologiques**

+ **Les données de séries chronologiques contiennent des observations d’une ou de plusieurs unités sur de nombreuses périodes**. 

+ **Disons que vous voulez comprendre la croissance démographique d’une ville au fil du temps. Vous mesureriez la population chaque année pour générer un ensemble de données contenant une colonne d’années et une colonne de chiffres de population.**

####**3. Que sont les objets xts ?**

+ Où le package xts entre-t-il en jeu? 

+ **xts facilite la manipulation des données de séries chronologiques dans R en produisant des objets xts, qui consistent en une matrice associée à un index temporel**. 

+ Ici, vous pouvez voir une matrice standard de valeurs numériques dans R - supposons qu’il s’agisse de données sur la population de la ville. 

+ Pour produire un objet xts à partir de cette matrice, il suffit d’attacher un vecteur de dates. 

+ Dans ce cas, nous joindrons des années correspondant aux mesures de population. 

+ Alors qu’une trame de données normale peut contenir une variable temporelle, un objet xts est indexé à temps, ce qui facilite la manipulation de manière intéressante. 

+ Par exemple, si nous voulons ajouter une nouvelle ligne de données au milieu de la période, xts sait placer cette ligne entre les dates les plus proches dans l’index de temps.

###**Enregistrement et exportation d’objets xts**


####**1. Enregistrement et exportation d’objets xts**

+ ***Une autre tâche de base dans la manipulation de séries chronologiques consiste à enregistrer vos données. Nous allons couvrir deux options pour enregistrer des données de séries chronologiques dans R.***

####**2. Enregistrer en tant que rds**

+ ***Tout d’abord, le moyen le plus efficace et le plus fiable de stocker vos objets xts est de les enregistrer en tant que fichiers rds à l’aide de la commande saveRDS().*** 

+ ***Lorsque vous rechargez des données de série chronologique dans R à partir d’un fichier rds à l’aide de readRDS(), les données conservent toutes leurs caractéristiques, y compris l’index temporel de l’objet xts***. 

+ ***Pour cette raison, il s’agit de la méthode préférée pour enregistrer des objets xts lorsque vous travaillez dans R.**

####**3. Enregistrement au format csv**

+ ***Mais que se passe-t-il si vous devez partager vos données de séries chronologiques avec un collègue qui n’utilise pas R ?*** 

+ ***Dans ce scénario, vous ne devez pas enregistrer vos données dans un fichier rds car votre collègue peut ne pas être en mesure de lire le fichier.*** 

+ ***Au lieu de cela, vous pouvez exporter votre objet xts vers un type de fichier facilement lisible, tel qu’un fichier de valeurs séparées par des virgules (ou CSV). ***

+ ***Pour ce faire, vous devez utiliser la commande write.zoo(), qui vous permet de spécifier le format de votre fichier d’exportation.*** 

+ ***Cependant, la compatibilité s’accompagne d’une perte d’informations. Lorsque vous rouvrez le fichier CSV à l’aide de read.zoo(), vous constaterez que vos données sont codées en tant qu’objet zoo plutôt qu’en tant qu’objet xts. Avant de poursuivre votre analyse, vous devez reconvertir l’objet en xts.***

####**1. Gestion des manquants**

+ ***L’un des défis les plus courants lorsque l’on travaille avec des données de séries chronologiques est le problème de l’absence.***

####**2. Absence**

+ **Si nos données contiennent un grand nombre de valeurs manquantes, il peut être difficile d’identifier des tendances dans les données et peut-être impossible d’effectuer une analyse statistique**. 

+ **Heureusement, xts et zoo offrent une poignée de commandes pour aider à remplir les données manquantes avec des valeurs pertinentes.**

####**3. Remplir les NA avec la dernière observation**


+ **La méthode la plus courante pour traiter les données manquantes consiste à utiliser la dernière observation reportée, ou LOCF**. 

+ **Cette approche comble les valeurs manquantes avec la valeur la plus récente disponible dans nos données.** 

+ **Par exemple, disons que nous avons des données de séries chronologiques annuelles jusqu’aux années 1980, mais qu’il nous manque des observations pour 1985 à 1987**. 

+ **La commande na-dot-locf() remplacera les NA pour ces années avec la dernière observation reportée, en l’occurrence les données de 1984.**

####**4. Remplir les NA avec l’observation suivante**

+ **Une technique similaire adopte l’approche opposée en utilisant l’observation suivante reportée en arrière, ou NOCB**. 

+ **Au lieu d’utiliser les données de 1984, cette technique prend nos données existantes de 1988 et les projette à rebours.**

####**5. Interpolation linéaire**

+ **Dans certains cas, il peut être plus judicieux de calculer une valeur quelque part entre vos points de données existants.**

+ **Cette approche, appelée interpolation linéaire, vous permet de calculer une nouvelle valeur entre vos points de données pondérés dans le temps**. 

+ Dans notre exemple, la valeur interpolée pour 1985 correspondra au quart du chemin entre nos données existantes de 1984 et nos données existantes de 1988. 

+ **La valeur interpolée pour 1986 sera la moitié du chemin entre nos points de données existants. La meilleure technique dépend du type de données que vous utilisez et des tendances que vous vous attendez à trouver.**

```r
# Add a quarterly difference in gdp
gdp$quarterly_diff <- diff(gdp$gdp, lag = 1, differences = 1)

# Split gdp$quarterly_diff into years
gdpchange_years <- split(gdp$quarterly_diff, f = "years")

# Use lapply to calculate the cumsum each year
gdpchange_ytd <- lapply(gdpchange_years, FUN = cumsum)

# Use do.call to rbind the results
gdpchange_xts <- do.call(rbind, gdpchange_ytd)

# Plot cumulative year-to-date change in GDP
plot.xts(gdpchange_xts, type = "h")
```

+ ***Dans l'exercice précédent, vous avez utilisé endpoints() et period.apply() pour calculer rapidement la moyenne des victoires/défaites des Red Sox de Boston à la fin de chaque saison. Mais que se passe-t-il si vous avez besoin de connaître la moyenne cumulée tout au long de chaque saison ? Les statisticiens et les amateurs de sport se fient souvent à cette moyenne pour comparer une équipe avec ses rivaux.***

+ **Pour calculer une moyenne cumulée à chaque saison, vous devrez revenir à la formule split-lapply-rbind pratiquée au chapitre trois. Tout d'abord, vous diviserez les données par saison, puis vous appliquerez une fonction de moyenne cumulative à la colonne win_loss à chaque saison, puis vous lierez les valeurs dans un objet xts.**

+ *Une fonction personnalisée cummean(), qui génère une somme cumulée et la divise par le nombre de valeurs incluses dans la somme, a été générée pour vous. Les données redsox_xts, y compris la colonne win_loss, sont disponibles dans votre espace de travail.*

####**Instructions**

+ Utilisez split() pour décomposer les données redsox_xts en saisons (dans ce cas, des années). Attribuez ceci à redsox_seasons.
+ Utilisez lapply() pour calculer la moyenne cumulée pour chaque saison. Pour cet exercice, une fonction cummean() a été conçue pour calculer la somme (en utilisant cumsum()) et la divise par le nombre d'entrées dans la somme (en utilisant seq_along()). Enregistrez ces données dans redsox_ytd.
+ Utilisez do.call() avec rbind pour convertir la sortie de votre liste en un seul objet xts (redsox_winloss) qui contient la moyenne des gains/pertes tout au long de chaque saison.
+ Utilisez plot.xts() pour afficher la moyenne cumulée des victoires/défaites au cours de la saison 2013. Laissez l'argument ylim tel quel dans votre code pré-écrit.



```r
cummean <- function (x){
    cumsum(x)/seq_along(x)
}
```

```r
# Split redsox_xts win_loss data into years 
redsox_seasons <- split(redsox_xts$win_loss, f = "years")

# Use lapply to calculate the cumulative mean for each season
redsox_ytd <- lapply(redsox_seasons, cummean)

# Use do.call to rbind the results
redsox_winloss <- do.call(rbind, redsox_ytd)

# Plot the win_loss average for the 2013 season
plot.xts(redsox_winloss["2013"], ylim = c(0, 1))
```

***Bien joué! Votre graphique montre la moyenne cumulée des victoires/défaites après chaque match de la saison 2013 ! Étant donné que chaque valeur est une moyenne cumulative au cours de la saison, il est logique que la valeur soit extrêmement volatile au début de la saison, lorsque chaque victoire ou défaite fait une plus grande différence.***
